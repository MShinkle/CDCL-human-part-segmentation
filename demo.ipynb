```python
from cdcl.wrapper import CDCL_process
from cdcl.viz import show_parts_overlay
from matplotlib import pyplot as plt

%matplotlib inline

process = CDCL_process()
image_filepath = "example_input.jpg"
img = plt.imread(image_filepath)
parts = process.extract_parts(img, target_size=[15,15], scales=[1.0], verbose=False, input_rgb=True)
show_parts_overlay(parts[0], image_filepath)
```
