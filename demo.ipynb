```python
from wrapper import compute_part_features

images_dir = "/hdd01/stimuli/LHimages/images/val"
part_features = compute_part_features(images_dir, scales=[1.0], target_size=15)
```

```python
from matplotlib import pyplot as plt
import os
%matplotlib inline
parts = ['faces', 'trunks', 'arms', 'hands', 'legs', 'feet']

image_i = 59
original_img = plt.imread(os.path.join(images_dir, f'im{image_i:07d}.png'))
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(original_img[...,:3], extent=[0, 1, 0, 1])
    ax.imshow(part_features[image_i, i], cmap='gray', alpha=.5, extent=[0, 1, 0, 1])
    ax.set_title(parts[i])
```
